```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
sns.set_style("darkgrid")
```

```python
# Specify the path to your local file
#file_path = r"D:\EOai4BIO\tables\stackedNormImage_noMarch_TRAINING_DATA.csv"
file_path = r"D:\EOai4BIO\tables\updated_stackedNormImage_TRAINING_DATA1.csv"

# Load the dataset
dataset = pd.read_csv(file_path)

# Display the first few rows of the dataset to verify it's loaded correctly
print(dataset.head())
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 25  # set width to 10 inches
fig_size[1] = 7   # set height to 6 inches
plt.rcParams["figure.figsize"] = fig_size
```

```python
# Create an array with the colors you want to use
custom_palette = ["#2A4000", "#E69800", "#8400A8", "#4CE600", "#FF73DF", "#FFFF00", "#00FFC5"]

# # Custom label-color mapping
# label_color_map = {
#     'Picea abeis': '#2A4000',        # Previously 'PiAb'
#     'Populus balsamifera': '#E69800', # Previously 'PoBa'
#     'Fraxinus excelsior': '#8400A8',  # Previously 'FrEx'
#     'Alnus incana': '#4CE600',        # Previously 'AlIn'
#     'Quercus robur': '#FF73DF',       # Previously 'QuRo'
#     'Alnus glutinosa': '#FFFF00',     # Previously 'AlGl'
#     'Salix alba': '#00FFC5'           # Previously 'SaAl'
# }

# Custom label-color mapping
label_color_map = {
    'Picea abies': '#2A4000',        # Previously 'PiAb'
    'Populus balsamifera': '#E69800', # Previously 'PoBa'
    'Fraxinus excelsior': '#8400A8',  # Previously 'FrEx'
    'Alnus incana': '#4CE600',        # Previously 'AlIn'
    'Quercus robur': '#FF73DF',       # Previously 'QuRo'
    'Alnus glutinosa': '#FFFF00',     # Previously 'AlGl'
    'Salix alba': '#00FFC5'           # Previously 'SaAl'
}
```

```python
#Edit the class labels
# Define your label mappings
label_map = {
    1: 'Picea abies',          # Corrected and updated name format
    2: 'Populus balsamifera',  # Updated name format
    3: 'Fraxinus excelsior',   # Updated name format
    4: 'Alnus incana',         # Updated name format
    5: 'Quercus robur',        # Updated name format
    6: 'Alnus glutinosa',      # Added missing mapping for label 6
    7: 'Alnus glutinosa',      # Updated name format
    8: 'Salix alba'            # Updated name format
}

# Map your original labels to custom labels
dataset['Tree Species'] = dataset['Label'].map(label_map)
```

```python
graph = sns.lineplot(x='band name',y='band values', hue = 'Tree Species', data=dataset, palette=label_color_map)
#sns.set_style("ticks")
# Rotate the x-axis labels to vertical
plt.xticks(rotation=90)  # Rotate labels to 90 degrees for vertical alignment
plt.xlim(left=-1)  # Adjust this value based on your data to add space at the beginning of the x-axis
# Change the color of the plot area to light gray
plt.gca().set_facecolor("#f4f4f4")

# Add a legend with italic labels
plt.legend(title='', 
           loc='upper left', 
           bbox_to_anchor=(1, 1), 
           frameon=False, 
           edgecolor='none', 
           prop={'style': 'italic', 'size': 14})  # Set text to italic and adjust the size


# Optionally, add labels, title, and grid
plt.xlabel('Bands and Dates',fontsize=14)
plt.ylabel('Reflectance values',fontsize=14)

print (graph)
plt.savefig('my_lineplot.png',dpi = 1000)
```
